```python
#W4A6
def nguyen_to(n):
  for i in  range(2, int(n)):
    if n%i == 0:
      return False
  return True
while True:
  a, b = list(map(int, input().split()))
  if b >= a and a>0 and b>0:
    break
  else:
    print("Nhập lại: ")
tong = 0
for i in range(a, b+1):
  if nguyen_to(i):
    tong += i
print(tong)
```

```python
#W4A7
def nguyen_to(n):
  for i in range(2, n):
    if n%2 == 0:
      return False
  return True
so = int(input())
danh_sach = []
if nguyen_to(so):
  print("Không có số nguyên tố")
else:
  for j in range(2,so):
    if so%j == 0:
      if nguyen_to(j):
        danh_sach.append(j)
  print(max(danh_sach))
```

```python
#W4A8
def dao_nguoc(n):
  return int(str(n)[::-1])
def palindrome(n):
  return int(n) == int(str(n)[::-1])
def thuc_hien():
  while True:
    n = int(input())
    if n > 0:
      break
    else:
      print("Nhập lại: ")
  buoc = 0
  hien_tai = n
  while not palindrome(hien_tai):
    so_nguoc = dao_nguoc(hien_tai)
    hien_tai = hien_tai + so_nguoc
    buoc += 1
  print(buoc)
  print(hien_tai)
if __name__ == "__main__":
  thuc_hien()
```

```python
#W4A9
import math
def khac_nhau(n):
  s = str(n)
  return len(s) == len(set(s))
def thuc_hanh():
  danh_sach = []
  n = int(input())
  for i in range(1, int(math.sqrt(n))+1):
    bac2 = i**2
    if khac_nhau(bac2):
      danh_sach.append(bac2)
  if danh_sach:
    print(' '.join(map(str, danh_sach)))
  else:
    print("no ")
if __name__ == "__main__":
  thuc_hanh()
```

```python
#W4A10
def collatz(n):
  if n == 1:
    return 1
  if n%2 == 0:
    return n//2
  else:
    return 3*n + 1
n = int(input())
danh_sach = {}
for i in range(1,n):
  dem = 1
  j = i
  while i > 1:
    i = collatz(i)
    dem += 1
  danh_sach[j] = dem
key_lon = max(danh_sach, key = danh_sach.get)
value_lon = danh_sach[key_lon]
print(key_lon, value_lon)
```

```python
#W4A13
def than_thiet(a, b):
  tong_a = 0
  tong_b = 0
  for i in range(1, a):
    if a%i == 0:
      tong_a += i
  for j  in range(1, b):
    if b%j == 0:
      tong_b += j
  if tong_a == b and tong_b == a:
    return True
  else:
    return False
x, y = list(map(int, input().split()))
than_thiet(x, y)
```
