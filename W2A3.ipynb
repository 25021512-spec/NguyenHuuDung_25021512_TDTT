```
a = int(input("Nhập số liệu đầu tiên: "))
b = int(input("Nhập số liệu thứ hai: "))
tong = a+b
hieu = a-b
tich = a*b
phan_nguyen = a//b
phan_du = a%b
chia_thuc = a/b
print("Tổng hai số là:", tong)
print("Hiệu hai số là:", hieu)
print("Tích hai số là:", tich)
print("Phần nguyên hai số là:", phan_nguyen)
print("Phần dư hai số là:", phan_du)
print("Chia thực hai số là:", round(chia_thuc, 2))
```
