```python
#W6A4
ds = input().split()
tu_dien = {}
for i in ds:
  k, v = i.split(':')
  if k not in tu_dien:
    tu_dien[k]= []
  tu_dien[k].append(v)
print(tu_dien)
```

```python
#W6A5
ds = list(map(int, input().split()))
tu_dien = {}
positive = []
negative = []
zero =[]
for i in ds:
  if i > 0:
    positive.append(i)
  elif i < 0:
    negative.append(i)
  else:
    zero.append(i)
tu_dien['positive'] = len(positive)
tu_dien['negative'] = len(negative)
tu_dien['zero'] = len(zero)
print(tu_dien)
```

```python
#W6A9
dict1 = eval(input())
dict2 = eval(input())
for k, v in dict2.items():
  if k in dict1:
    dict1[k] += v
  else:
    dict1[k] = v
print(dict(sorted(dict1.items())))
```

```python
#W6A10
so_nguyen = list(map(int, input().split()))
k = int(input())
danh_sach= []
cac_cap = set()
for i in range(len(so_nguyen)):
  for j in range(i+1, len(so_nguyen)):
    if so_nguyen[i] + so_nguyen[j] == k:
      if (so_nguyen[i], so_nguyen[j]) not in cac_cap:
        danh_sach.append((so_nguyen[i], so_nguyen[j]))
        cac_cap.add((so_nguyen[i], so_nguyen[j]))
print(danh_sach)
```

```python
#W6A12
ds = list(map(int, input().split()))
dem = {}
for i in ds:
  dem[i] = dem.get(i, 0) + 1
lon_nhat = max(dem.values())
danh_sach = []
for k, v in dem.items():
  if v == lon_nhat:
    danh_sach.append(k)
print(min(danh_sach))
```

```python
#W6A14
d1 = input().split()
d2 = input().split()
d11 = []
d22 = []
for i in d1:
  if i not in d11:
    d11.append(i)
for j in d2:
  if j not in d22:
    d22.append(j)
ds = []
for k in d11:
  if k in d22:
    ds.append(int(k))
ds.sort()
print(ds)
```

```python
#W6A15
dict0 = input().split()
k = int(input())
dict1 = {}
bien = [dict0[i] for i in range(0, len(dict0), 2)]
gt = [int(dict0[j]) for j in range(1, len(dict0), 2)]
for b, g in zip(bien, gt):
  if g > k:
    dict1[b] = g
print(dict1)
```
