```python
#W6A20
def dau_tien(lst, k):
  for i in lst:
    if i == k:
      return lst.index(k)+1
  return -1
lst = list(map(int, input().split()))
k = int(input())
res = dau_tien(lst, k)
print(res)
```

```python
#W6A23
n = list(map(int, input().split()))
x = int(input())
ds = []
for i in n:
  if i >= x:
    if i not in ds:
      ds.append(i)
print(sorted(ds))
```

```python
#W6A24
n = list(map(int, input().split()))
for i in range(len(n)):
  if n[i] == 0:
    n.pop(i)
    n.insert(0, 0)
print(*n)
```

```python
#W6A26
n = list(map(int, input().split()))
tong = 0
for i in range(1, len(n)-1):
  if n[i-1] <= n[i] >= n[i+1]:
    tong += n[i]
print(tong)
```

```python
#W6A28
m = list(map(int, input().split()))
if m == sorted(m) and len(set(m)) == len(m):
  print("Dãy tăng thực sự")
elif m == sorted(m, reverse=True) and len(set(m)) == len(m):
  print("Dãy giảm thực sự")
else:
  print("Vô danh")
```
