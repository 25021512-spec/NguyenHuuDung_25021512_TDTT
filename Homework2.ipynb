```python
#Bài 3
import math
canh = list(map(float, input("Nhập 3 cạnh tam giác: ").split()))
a = canh[0]
b = canh[1]
c = canh[2]
p = (a+b+c)/2
if len(canh)<3 or len(canh)>3:
  print("Số liệu lỗi")
else:
  if a+b<=c or a+c<=b or b+c<=a:
    print("Khong la ba canh cua mot tam giac")
  elif a==b==c:
    print("Tam giac deu")
    print(f"Dien tich tam giac la: {math.sqrt(p*(p-a)*(p-b)*(p-c))}")
  elif a**2+b**2==c**2 or a**2+c**2==b**2 or b**2+c**2==a**2:
    print("Tam giac vuong")
    print(f"Dien tich tam giac la: {math.sqrt(p*(p-a)*(p-b)*(p-c))}")
  elif a==b or a==c or b==c:
    print("Tam giac can")
    print(f"Dien tich tam giac la: {math.sqrt(p*(p-a)*(p-b)*(p-c))}")
  else:
    print("Tam giac thuong")
    print(f"Dien tich tam giac la: {math.sqrt(p*(p-a)*(p-b)*(p-c))}")
```

```python
#Bài 5
a, b, c = list(map(float, input("Nhập ba số a,b và c: ").split()))
delta = b**2-4*a*c
print(f"Phương trình bậc hai: {a}x^2 + {b}x + {c} = 0")
if delta < 0:
  print("Phương trình vô nghiệm")
elif delta == 0:
  print(f"Phương trình có nghiệm kép x1 = x2 = {-b/(2*a)}")
else:
  print(f"Phương trình có hai nghiệm phân biệt: x1 = {(-b+delta**0.5)/(2*a)} và x2 = {(-b-delta**0.5)/(2*a)}")
```

```python
#Bài 6
so_sanh = list(map(float, input().split()))
if not so_sanh:
  print("Lỗi số")
else:
  max = so_sanh[0]
  for i in so_sanh:
    if i > max:
      max = i
  print("Số lớn nhất là: ",max)
```

```python
#Bài 8
a, b, c, d, m, n = list(map(float, input().split()))
print(f"{a}x + {b}y = {m}")
print(f"{c}x + {d}y = {n}")
print("giá trị của x =", (m*d-b*n)/(a*d-b*c))
print("giá trị của y =", (a*n-m*c)/(a*d-b*c))
```

```python
#Bài 9
giay = int(input("Nhập số giây: "))
gio = giay/3600
hour = gio - int(gio)
if hour == 0:
  print(f"Giờ làm việc là {int(gio)}h")
else:
  phut = hour*60
  min = phut - int(phut)
  if min == 0:
    print(f"Thời gian làm việc: {int(gio)}h {phut}m")
  else:
    sec = giay - int(gio)*3600 - int(phut)*60
    print(f"Thời gian làm việc: {int(gio)}h {int(phut)}m {sec}s")
```
