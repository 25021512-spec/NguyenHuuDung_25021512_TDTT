```python
#W5A1
def Max_Of_Two(a, b) -> int:
  return max(a, b)
def thuc_hanh():
  n = int(input("Nhập n: "))
  m = int(input("Nhập m: "))
  return Max_Of_Two(n, m)
if __name__ == "__main__":
  print(thuc_hanh())
```

```python
#W5A2
def swap(a, b) -> tuple[int, int]:
  return b, a
def thuc_hanh():
  n = int(input("Nhập n: "))
  m = int(input("Nhập m: "))
  return swap(n, m)
if __name__ == "__main__":
  print(thuc_hanh())
```

```python
#W5A3
def nguyen_to(n):
  if n < 2:
    return False
  for i in range(2, n):
    if n % i == 0:
      return False
  return True
n = int(input())
for i in range(2, n + 1):
  if nguyen_to(i):
    print(i, end=" ")
```

```python
#W5A4
def pefnum(n):
  sum = 0
  for i in range(1, n):
    if n%i == 0:
      sum = sum + i
  if sum == n:
    return True
  else:
    return False
if __name__ == "__main__":
  n = int(input())
  res = pefnum(n)
  if res:
    print(f"{n} là số hoàn hảo")
  else:
    print(f"{n} không là số hoàn hảo")
```

```python
#W5A5
def lin(n, k):
  for i in range(len(n)):
    if n[i] == k:
      return i+1
  return -1
if __name__ == "__main__":
  x = list(input().split())
  y = input()
  res = lin(x, y)
  print(res)
```

```python
#W5A6
def giai_thua(n):
  if n == 0 or n == 1:
    return 1
  else:
    return n*giai_thua(n-1)
x = int(input())
print(giai_thua(x))
```

```python
#W5A7
def may_tinh(phep_toan, so1, so2):
  if phep_toan == '+':
    return so1 + so2
  elif phep_toan == '-':
    return so1 - so2
  elif phep_toan == '*':
    return so1 * so2
  elif phep_toan == '/':
    return so1 / so2
  else:
    return "Lỗi"
if __name__ =="__main__":
  num1 = float(input())
  operate = input("Nhập phép toán tử: ")
  num2 = float(input())
  res = may_tinh(operate, num1, num2)
  print(res)
```

```python
#W5A10
def dang_cau(a, b):
  ax_a_b = {}
  ax_b_a = {}
  for chuoi_a, chuoi_b in zip(a, b):
    if chuoi_a in ax_a_b:
      if ax_a_b[chuoi_a] != chuoi_b:
        return False
    elif chuoi_b in ax_b_a:
      if ax_b_a[chuoi_b] != chuoi_b:
        return False
    else:
      ax_a_b[chuoi_a] = chuoi_b
      ax_b_a[chuoi_b] = chuoi_a
  return True
if __name__ == "__main__":
  a = input()
  b = input()
  res = dang_cau(a, b)
  print(res)
```
