```python
#bài 4
h = int(input())
for i in range (1,h+1):
  if i == 1:
    sao = 1
    cach = h-1
    print(" " * cach + "*" * sao + " " * cach)
  elif 1<i<h:
    cach_ben = h - i
    cach_giua = 2*i-3
    print(" " * cach_ben + "*" + " " * cach_giua + "*" + " " * cach_ben)
  else:
    print("*" * (2*h-1))
```

```python
#bài 16
n = int(input())
chan = 0
le = 0
while n>0:
  chu_so = n%10
  if chu_so%2==0:
    chan += 1
  else:
    le += 1
  n = (n-chu_so)/10
print(f"{chan}, {le}")
```
